```python
import pandas as pd
import statsmodels.api as sm
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_absolute_error
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Carregamento 
# Load Data
df = pd.read_excel('1st_Phase_-_Selective_Process_-_Data_Science_-_DataBase.xlsx')

# Tira o ID, pois não tem valor preditivo 
# # Remove ID, because it doesn't have predictive value
df_analise = df.drop('ID', axis=1)
```

```python
# Dummificação - transforma os dados categoricos em números
# O parâmetro drop_first=True evita a multicolinearidade (se não é Capital,é Interior, .: não precisamos de duas colunas).
 
# #Dummification - converts categorical text into numerical values.
#The parameter drop_first=True avoids multicollinearity (if it’s not Capital, it’s Interior, so we don’t need two columns).
df_modelo = pd.get_dummies(df_analise, columns=['EDUCAÇÃO', 'CARGO', 'LOCAL'], drop_first=True)
```

```python
# Definição das Variáveis do Modelo: Result é o alvo da predição; Vars são as variáveis usadas para prever o alvo
# Model Variable Definition: 'Result' is the prediction target; 'Vars' are the features used to predict the target.
Result = df_modelo['SALARIO_MENSAL']
Vars = df_modelo.drop('SALARIO_MENSAL', axis=1)
```

```python
# A inclusão da constante evita que a linha de regressão seja forçada a cruzar a origem (0,0), 
# permitindo que o modelo capture o valor inicial (viés) dos dados.
# Including the constant prevents the regression line from being forced through the origin (0,0),
# allowing the model to capture the baseline value (bias) of the data.var = sm.add_constant(Vars)
var = sm.add_constant(Vars)
```

```python
# Separação em Treino (80%) e Teste (20%) / Train-Test Split (80/20)
X_train, X_test, y_train, y_test = train_test_split(Vars, Result, test_size=0.2, random_state=42)
```

```python
# Treinamento do Modelo OLS / OLS Model Training
# Treinamos apenas com os dados de treino / Trained only on training data
modelo = sm.OLS(y_train, X_train.astype(float)).fit()
```

```python
# Previsões e Métricas / Predictions and Metrics
y_pred_test = modelo.predict(X_test.astype(float))
print("=====MÉTRICAS DE VALIDAÇÃO / VALIDATION METRICS====")
print(f"R² (Teste/Test): {r2_score(y_test, y_pred_test):.4f}")
print(f"MAE (Teste/Test): R$ {mean_absolute_error(y_test, y_pred_test):.2f}")

# =====MÉTRICAS DE VALIDAÇÃO / VALIDATION METRICS====
# R² (Teste/Test): 0.3491 O 
# MAE (Teste/Test): R$ 867.35

##Sobre R² / About R²
# O quanto o modelo explica - aprox. 35% / How much the model explains - about 35%
# Isso é normal para amostras pequenas / This is normal for small samples

##Sobre MAE / About MAE
# Erro absoluto do modelo - 15% (867 / média=5777) / Abolute Error of the model - 15% (867 / mean=5777)

```

The analysis concludes that the company is already in full compliance with the new policy regarding non-discrimination by location. Although a superficial review of raw averages might suggest certain disparities, the multiple regression model reveals that location is not a determining factor for compensation. The most critical data point supporting this finding is the p-value of 0.109 for the "Interior" variable. Since this figure exceeds the standard threshold of 0.05, there is no statistical evidence to suggest that being based in the interior negatively impacts an employee's salary; mathematically, any observed variation in this variable is considered "noise" rather than a corporate practice.

The results further demonstrate that compensation within the organization is actually driven by merit-based factors and seniority. Age (p = 0.000) emerged as the strongest predictor, indicating that accumulated experience is highly valued. Furthermore, holding a Management position (p = 0.027) or possessing Higher Education (p = 0.017) results in statistically significant salary increases. These findings confirm that the company's pay structure is rooted in professional qualifications rather than geographic bias.

Finally, it is worth noting a key insight from the descriptive analysis: the distribution of roles plays a significant role in the overall perception of pay gaps. Currently, 100% of the executive directors are based in the capital. Given that these positions command the highest salaries, they inevitably skew the city's average salary upward. Understanding this structural concentration is essential to avoid misinterpreting the broader data set.

Conclusão: a empresa já está a cumprir a nova política de não discriminação salarial por localização.  
Embora uma análise superficial das médias possa sugerir disparidades, o modelo estatístico (regressão múltipla) revela que a localização não é um fator determinante para o salário. O dado mais importante para esta conclusão é o p-valor de 0.109 para a variável LOCAL_INTERIOR. Como este valor é superior ao limite padrão de 0.05, não existe evidência estatística de que o facto de um funcionário estar no interior impacte negativamente a sua remuneração; matematicamente, qualquer variação observada nesta variável é considerada ruído e não uma prática da emprea.  
Os resultados demonstram que o que realmente define os salários na organização são fatores de mérito e senioridade. A Idade (p = 0.000) aparece como o preditor mais forte, indicando que a experiência acumulada é valorizada. Possuir um Cargo de Gerente (p = 0.027) ou Educação Superior (p = 0.017) garante aumentos salariais estatisticamente comprovados.
Acho importante comentar quem durante a análise descritiva dos dados, percebi que a distribuição dos cargos pode ser um fator relevante: todos os diretores estão na capital. Como esses cargos tem salários mais altos, eles vão afetar dados como a média do salário na cidade.
